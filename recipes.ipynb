```python
import numpy as np
import pandas as pd
```

```python
recipes = pd.read_json('clean_cookies_recipes.json')
```

```python
recipes
```

```python
title_to_ingredients = {}
for index, row in recipes.iterrows():
    title_to_ingredients[row['title']] = row['recipe']
```

```python
def find_recipes(ingredients):
    threshold = 0.9
    matches = []
    for key, value in title_to_ingredients.items():
        num_ingr_in_recipe = len(value)
        if num_ingr_in_recipe > 1:
            matching_ingredient = 0.0
            for ingr in value:
                for ingr_disp in ingredients:
                    if ingr == ingr_disp:
                        matching_ingredient += 1
            if matching_ingredient / num_ingr_in_recipe >= threshold:
                matches.append(key)
    return matches
```

```python
fake_ingredients = ['butter', 'tomato', 'carrot', 'onion', 'water', 'salt', 'olive oil', 'egg']
```

```python
find_recipes(fake_ingredients)
```
